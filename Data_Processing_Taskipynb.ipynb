```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
```

```python
df['age'] = df['age'].fillna(df['age'].median())
```

```python
df
```

```python
df['city'] = df['city'].fillna(df['city'].mode()[0])
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
df = pd.get_dummies(df, columns=['city_Mumbai'])
df
```

```python
le = LabelEncoder()
```

```python
df['gender'] = le.fit_transform(df['gender'])
df
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
num_cols = ['age', 'annual_income']
minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[num_cols] = minmax.fit_transform(df[num_cols])
print("\nMin-Max Scaler Data:\n")
print(df_minmax.head())
```

```python
standard = StandardScaler()
df_standard = df.copy()
df_standard[num_cols] = standard.fit_transform(df[num_cols])
print("\nStandardized Data:\n")
print(df_standard.head())
```
